# **Código de verificaçãoo de barras.**



Códigos de barras repetidos em produtos diferentes são proibidos por lei e geram problemas sérios para a gestão e vendas, pois cada produto deve ter um código exclusivo. A duplicação ocorre por falha humana no cadastro (copiar um código similar) ou por sistemas que não bloqueiam a duplicidade. Isso causa conflitos na gestão de estoque, distribuição, rastreamento e vendas. A solução é corrigir os cadastros, garantindo que cada variação de produto (cor, tamanho, pacote) tenha um código único.

Mas não existe uma lei **federal específica** no Brasil que proíba explicitamente o uso de códigos de barras repetidos para produtos diferentes. A proibição ou a exigência de unicidade decorrem de normas técnicas e implicações legais indiretas, especialmente relacionadas ao **Código de Defesa do Consumidor (CDC) e à legislação tributária.**

Código de Defesa do Consumidor (CDC) - Lei nº 8.078/90

A pedido do nosso cliente vamos verificar se a incoenrencia ou repetições nos codigos de barra de ambos inventarios.

## Verificação feita no inventario de 31-12-2022

```python
import pandas as pd

# Atribua df_dez_22 a df_cleaned para esta análise
df_cleaned = df_dez_22

# verificação de códigos repetidos
def encontrar_conflitos_simples(dados):
    """
    Identifica códigos de barra que possuem mais de um nome de produto associado.

    A entrada deve ser uma lista de tuplas/listas no formato:
    [ (codigo_barra, nome_produto), ... ]
    """

    mapa_codigos_nomes = {}

    # 1. Mapear cada código de barra a um conjunto (set) de nomes
    for codigo, nome in dados:
        # Usamos .setdefault para iniciar o conjunto se o código for novo
        # e .add para incluir o nome (sets garantem nomes únicos)
        mapa_codigos_nomes.setdefault(codigo, set()).add(nome)

    conflitos = {}

    # 2. Verificar quais códigos têm mais de um nome associado (conflito)
    for codigo, nomes_associados in mapa_codigos_nomes.items():
        if len(nomes_associados) > 1:
            conflitos[codigo] = list(nomes_associados)

    return conflitos

# Use o DataFrame df_cleaned e converta-o para uma lista de tuplas
dados_de_entrada = list(zip(df_cleaned['Código'], df_cleaned['Descrição']))

# Chama a função
conflitos_encontrados = encontrar_conflitos_simples(dados_de_entrada)

# Exibe o resultado
print("--- Relatório de Conflitos de Nomes ---")

if conflitos_encontrados:
    print(f" CÓDIGO(S) EM CONFLITO: {len(conflitos_encontrados)}")
    for codigo, nomes in conflitos_encontrados.items():
        print(f"\nCÓDIGO: **{codigo}**")
        print(f"   Nomes Diferentes Encontrados: {', '.join(nomes)}")
else:
    print(" NENHUM código de barra repetido foi associado a nomes diferentes.")
```

# **Analise de lucro, prejuízo e precificação.**

A análise de precificação e custo tem como objetivo verificar se os produtos estão sendo vendidos com lucro ou prejuízo, comparando o preço de venda com o custo médio de cada item.

Ela permite identificar falhas de precificação, ajustar preços e avaliar a rentabilidade por categoria.
Com isso, a empresa pode entender quais produtos geram mais lucro, quais devem ser reajustados e qual é o lucro potencial total do estoque.

Essa análise é essencial para evitar perdas financeiras, melhorar o planejamento de vendas e tomar decisões estratégicas mais seguras.

## Inventario de 2022

```python
import pandas as pd

# Certifique-se de que o DataFrame df_dez_22 esteja disponível e atribuído a 'df'
df = df_dez_22.copy() # Usamos uma cópia para não alterar o DataFrame original caso necessário

# Extras (indicadores rápidos)
valor_estoque = (df["Estoque"] * df["Custo Médio"]).sum()
valor_venda = (df["Estoque"] * df["Preço Venda"]).sum()
lucro_potencial = valor_venda - valor_estoque

print("\n Indicadores Gerais:")
print(f"Valor total de estoque (custo): R$ {valor_estoque:,.2f}")
print(f"Valor potencial de venda: R$ {valor_venda:,.2f}")
print(f"Lucro potencial: R$ {lucro_potencial:,.2f}")
```

## Inventario de 2023

```python
import pandas as pd

# Certifique-se de que o DataFrame df_dez_23 esteja disponível e atribuído a 'df'
df = df_dez_23.copy() # Usamos uma cópia para não alterar o DataFrame original caso necessário

# Calcular 'Custo Médio' para o df_dez_23
# A descrição indica 'Estoque Compra' é o valor total de custo e 'Estoque' é a quantidade.
# Então, Custo Médio = Estoque Compra / Estoque
# Lidar com divisões por zero, definindo o custo médio como 0 onde o estoque é 0.
df['Custo Médio'] = df.apply(
    lambda row: row['Estoque Compra'] / row['Estoque'] if row['Estoque'] > 0 else 0,
    axis=1
)

# Extras (indicadores rápidos)
valor_estoque = (df["Estoque"] * df["Custo Médio"]).sum()
valor_venda = (df["Estoque"] * df["Preço Venda"]).sum()
lucro_potencial = valor_venda - valor_estoque

print("\n Indicadores Gerais:")
print(f"Valor total de estoque (custo): R$ {valor_estoque:,.2f}")
print(f"Valor potencial de venda: R$ {valor_venda:,.2f}")
print(f"Lucro potencial: R$ {lucro_potencial:,.2f}")
```

## Conclusão

A comparação de Lucro e Prejuízo aponta para uma tendência positiva: a redução na probabilidade de prejuízo total do portfólio demonstra que as ações corretivas tiveram êxito em estabilizar a rentabilidade. Em suma, a empresa realizou uma transição bem-sucedida de uma gestão reativa (2022, focada na correção de falhas de data quality) para uma gestão mais proativa e robusta (2023, baseada em processos aprimorados), resultando em um inventário financeiramente mais estável e previsível.

# **Análise de Monte Carlo**

A análise de Monte Carlo simula diversos cenários possíveis de variação nos preços de venda e custos médios dos produtos para estimar o lucro total do estoque e o risco de prejuízo.
Com isso, é possível entender como as incertezas do mercado (como mudanças de preço ou custo) afetam os resultados da empresa, ajudando na tomada de decisão e no controle financeiro.




## Análise de Monte Carlo para o Inventário de 2022



```python
import matplotlib.pyplot as plt
import numpy as np

# Certifique-se de que o DataFrame df_dez_22 esteja disponível
df = df_dez_22.copy() # Usamos uma cópia para não alterar o DataFrame original caso necessário

# ====================================================================
# ANÁLISE DE MONTE CARLO (2022)
# ====================================================================

# Configurações da simulação
n_simulations = 2000  # Número de simulações a serem executadas
volatility_price_venda = 0.10  # 10% de variação no preço de venda
volatility_custo_medio = 0.08  # 8% de variação no custo médio

total_profit_22 = []

for _ in range(n_simulations):
    # Gerar variações aleatórias para Preço Venda e Custo Médio
    # Usamos uma distribuição normal com média 1 e desvio padrão igual à volatilidade
    # Multiplicamos pelo valor original para obter a variação
    variation_price_venda = np.random.normal(1, volatility_price_venda, len(df))
    variation_custo_medio = np.random.normal(1, volatility_custo_medio, len(df))

    # Calcular Preço Venda e Custo Médio simulados
    simulated_price_venda = df['Preço Venda'] * variation_price_venda
    simulated_custo_medio = df['Custo Médio'] * variation_custo_medio

    # Ajustar para não ter preços ou custos negativos
    simulated_price_venda[simulated_price_venda < 0] = 0
    simulated_custo_medio[simulated_custo_medio < 0] = 0

    # Considerando os produtos com Preço Venda original de 0 como erro, mantemos 0
    simulated_price_venda[df['Preço Venda'] == 0] = 0

    # Considerando os produtos com Custo Médio original de 0 como custo já amortizado, mantemos 0
    simulated_custo_medio[df['Custo Médio'] == 0] = 0

    # Calcular o lucro para cada item nesta simulação
    item_profit = (simulated_price_venda - simulated_custo_medio) * df['Estoque']

    # Somar os lucros dos itens para obter o lucro total do estoque na simulação
    total_profit_22.append(item_profit.sum())

# Converter para array numpy para facilitar cálculos
total_profit_22 = np.array(total_profit_22)

# Calcular métricas para 2022
mean_profit_22 = np.mean(total_profit_22)
VaR_95_22 = np.percentile(total_profit_22, 5) # Valor em Risco nos 5% piores cenários
prob_negative_22 = (total_profit_22 < 0).sum() / n_simulations
prob_success_22 = 1 - prob_negative_22

# ====================================================================
# VISUALIZAÇÃO DA DISTRIBUIÇÃO DE LUCRO (2022)
# ====================================================================
plt.figure(figsize=(12, 7))
N_22, bins_22, patches_22 = plt.hist(total_profit_22, bins=30, edgecolor='black', alpha=0.7)

# Destaca a área de Lucro (Verde) e Prejuízo (Vermelho)
for p in patches_22:
    if p.get_x() < 0:
        p.set_facecolor('red')
    else:
        p.set_facecolor('green')

# Linhas de referência
plt.axvline(0, color='black', linestyle='--', linewidth=2, label='Ponto de Equilíbrio (Lucro = 0)')
plt.axvline(mean_profit_22, color='blue', linestyle='-', linewidth=2, label=f'Lucro Esperado (Média): R$ {mean_profit_22:,.2f}')
plt.axvline(VaR_95_22, color='orange', linestyle=':', linewidth=2, label=f'VaR (95%): R$ {VaR_95_22:,.2f}')

# Anotação de Porcentagem de Sucesso
plt.text(total_profit_22.max() * 0.95, N_22.max() * 0.9,
         f'Probabilidade de Lucro: {prob_success_22:.2%}',
         horizontalalignment='right', fontsize=12,
         bbox=dict(facecolor='white', alpha=0.8, boxstyle='round'))

# Configurações Finais
plt.title("Distribuição do Lucro Total Estocado (Simulação Monte Carlo - 2022)")
plt.xlabel("Lucro Total (R$)")
plt.ylabel("Frequência de Simulações")
plt.legend(loc='upper left')
plt.grid(axis='y', alpha=0.5)
plt.tight_layout()
plt.show()

print("\nConclusão da Análise de Monte Carlo para o Inventário de 2022:")
print("A simulação de Monte Carlo para o inventário de 2022 revela insights importantes sobre o potencial de lucro e risco do portfólio.")
print(f"O lucro médio esperado é de R$ {mean_profit_22:,.2f}, com uma probabilidade de prejuízo total de {prob_negative_22:.2%}.")
print(f"O Valor em Risco (VaR 95%) é de R$ {VaR_95_22:,.2f}, indicando a perda máxima esperada nos 5% piores cenários.")
```
